```python
# load plotting libraries
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap

# module of routines for retrieving data
# and plotting on map
import wUMapPlots as Map
```

```python
# simply plot positions of available stations on map
Map.plotOnMap()
```

```python
# make contour plot of a variable (no background or lon-lat projection)

# load list of station codes
stations = Map.getStationList()

# get corresponding lon and lat coordinates
lon, lat = Map.getStationLonLat(stations)

# plot daily min Temperature for November 7, 2007
Tmax_nov_7_2007 = Map.loadDailyVariable(stations, '2007-11-07', 'TempMin')
Map.contourPlot(lon, lat, Tmax_nov_7_2007, title='Daily minimum temp, 2007-11-07')

# plot daily max Temperature for November 7, 2007
Tmax_nov_7_2007 = Map.loadDailyVariable(stations, '2007-11-07', 'TempMax')
Map.contourPlot(lon, lat, Tmax_nov_7_2007, title='Daily maximum temp, 2007-11-07')
```

```python
# make contour plot of a variable (lon-lat projection with map background)

# load list of station codes
stations = Map.getStationList()

# get corresponding lon and lat coordinates
lon, lat = Map.getStationLonLat(stations)

# plot daily min Temperature for November 7, 2007
Tmax_nov_7_2007 = Map.loadDailyVariable(stations, '2007-11-07', 'TempMin')
Map.contourPlotOnMap(lon, lat, Tmax_nov_7_2007, title='Daily minimum temp, 2007-11-07')

# plot daily max Temperature for November 7, 2007
Tmax_nov_7_2007 = Map.loadDailyVariable(stations, '2007-11-07', 'TempMax')
Map.contourPlotOnMap(lon, lat, Tmax_nov_7_2007, title='Daily maximum temp, 2007-11-07')
```

```python
# plot Humidity on map
Map.contourPlotVarOnMap('HumidityMean', '2007-11-07', npts=30, ncntrs=20)
```

```python
# plot time mean on map
# Humidity
Map.contourPlotMeanVarOnMap('HumidityMean', '2006-01-01', '2007-12-31', npts=30, ncntrs=20)

# Temperature
Map.contourPlotMeanVarOnMap('TempMax', '2006-01-01', '2007-12-31', npts=30, ncntrs=20)
Map.contourPlotMeanVarOnMap('TempMin', '2006-01-01', '2007-12-31', npts=30, ncntrs=20)
Map.contourPlotMeanVarOnMap('TempMean', '2006-01-01', '2007-12-31', npts=30, ncntrs=20)

# SLP
Map.contourPlotMeanVarOnMap('PressMax', '2006-01-01', '2007-12-31', npts=30, ncntrs=20)
```

```python
# plot wind vector field
Map.plotWindVectorsOnMap('2007-11-07')

# mean wind
Map.plotMeanWindVectorsOnMap('2006-01-01', '2007-12-31')
```

```python
# plot advection of Temperature
Map.plotAdvectionOnMap('CYYZ','TempMax','2007-11-07',width_fac=32,height_fac=25)
```
