```python
import wUPCA as PCA
import wURegression as Reg
import wUMapPlots as Map
import wUUtils as Util
```

```python
stations = Util.getStationList()
features = ['TempMax', 'HumidityMean', 'isFoggy']
ncomp = [6, 11, 5]

pcaData, transform_params = PCA.pcaConvert(stations, features, '2005-01-01', '2010-12-31', ncomp)
print('number of features: %d\n' % len(pcaData))
for ii in range(len(pcaData)):
     print('number of PC in feature %s: %d, number of values: %d' % (features[ii], len(pcaData[ii]), len(pcaData[ii][0])))
     print('  shape of each PC rotation: %s\n' % str(transform_params['pcas'][ii].components_.shape))
```

```python
import matplotlib.pyplot as plt

# plot principal components of maximum temperature
fig = plt.figure()
date_list = Reg.dateList('2005-01-01','2010-12-31')
plt.plot(date_list, pcaData[0][0], color='blue')
plt.plot(date_list, pcaData[0][1], color='cyan')
plt.plot(date_list, pcaData[0][2], color='black')
plt.plot(date_list, pcaData[0][3], color='red')
plt.plot(date_list, pcaData[0][4], color='green')
fig.autofmt_xdate()
plt.show()

# smoothed, zoomed plots
fig = plt.figure()
# PC 0
plt.plot(date_list[:100], pcaData[0][0][:100], color='black',linestyle='--')
plt.plot(date_list[:100], Reg.smooth(pcaData[0][0])[:100], color='blue')
plt.plot(date_list[:100], Reg.smooth(pcaData[0][0],5)[:100], color='red')
plt.plot(date_list[:100], Reg.smooth(pcaData[0][0],13)[:100], color='black')
# PC 1
plt.plot(date_list[:100], pcaData[0][1][:100], color='black', linewidth=3, linestyle='--')
plt.plot(date_list[:100], Reg.smooth(pcaData[0][1],13)[:100], color='cyan', linewidth=3)
# PC 5
plt.plot(date_list[:100], pcaData[0][5][:100], color='black', linewidth=2, linestyle='--')
plt.plot(date_list[:100], Reg.smooth(pcaData[0][5],13)[:100], color='green', linewidth=3)
fig.autofmt_xdate()
plt.show()
```

```python
# plot principal components of Humidity
fig = plt.figure()
plt.plot(date_list, pcaData[1][0], color='blue')
plt.plot(date_list, pcaData[1][1], color='cyan')
plt.plot(date_list, pcaData[1][2], color='black')
plt.plot(date_list, pcaData[1][3], color='red')
plt.plot(date_list, pcaData[1][4], color='green')
fig.autofmt_xdate()
plt.show()
```

```python
# variance explained by PC
print('Temp PC: \t' + str(transform_params['pcas'][0].explained_variance_ratio_))
print('Humidity PC: \t' + str(transform_params['pcas'][1].explained_variance_ratio_))

# components of the first 5 temperature PC
TempPC = transform_params['pcas'][0].components_
print('station\t PC1 \t\t PC2 \t\t PC3 \t\t PC4 \t\t PC5')
for ii in range(len(stations)):
    print('%s \t %f \t %f \t %f \t %f \t %f' \
             % (stations[ii] , TempPC[0,ii], TempPC[1,ii], TempPC[2,ii], TempPC[3,ii], TempPC[4,ii]))
```

```python
# plot PC on map
reload(Map)

# get longitude and latitude of stations 
lon, lat = Util.getStationLonLat(stations)

# get variance explained by each PC
vrncs = transform_params['pcas'][0].explained_variance_ratio_

# first PC is very homogeneous so pattern not interesting
Map.contourPlotOnMap(lon, lat, TempPC[0] , width_fac=18, height_fac=15, \
                     title = 'TempPC 0: %.1f%% of variance' % (100*vrncs[0]))
# second through fifth temperature PC:
Map.contourPlotOnMap(lon, lat, TempPC[1] , width_fac=18, height_fac=15, \
                     title = 'TempPC 1: %.1f%% of variance' % (100*vrncs[1]))
Map.contourPlotOnMap(lon, lat, TempPC[2] , width_fac=18, height_fac=15, \
                     title = 'TempPC 2: %.1f%% of variance' % (100*vrncs[2]))
Map.contourPlotOnMap(lon, lat, TempPC[3] , width_fac=18, height_fac=15, \
                     title = 'TempPC 3: %.1f%% of variance' % (100*vrncs[3]))
Map.contourPlotOnMap(lon, lat, TempPC[4] , width_fac=18, height_fac=15, \
                     title = 'TempPC 4: %.1f%% of variance' % (100*vrncs[4]))
Map.contourPlotOnMap(lon, lat, TempPC[28] , width_fac=18, height_fac=15, \
                     title = 'TempPC 28: %.1f%% of variance' % (100*vrncs[28]))
```
