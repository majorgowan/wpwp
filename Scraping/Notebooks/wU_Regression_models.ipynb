```python
# using data from multiple cities on multiple days

# get list of all available stations
import wUMapPlots as Map
stations = Map.getStationList()
# move target station to front
targStation = 'CYYZ'
stations = [targStation] + [s for s in stations if s != targStation]
# build model
featureData, target, model_params = Reg.multiCityTaylorModel(stations,'2005-01-01','2010-12-31',\
                                                           ['TempMax','dailyTempRange',\
                                                           'PressMax','dailyPressRange',\
                                                           'HumidityMean',\
                                                           'isWesterly','isSoutherly',\
                                                           'isFoggy'],targetVar='TempMax',lag=1,order=1,scale=True,verbose=False)
```

```python
# predict
date_list, pred, target, model_perf = Reg.multiCityTaylorPredict(model_params, \
                                                                 '2011-01-01', '2012-12-31', \
                                                                 actual=True)
Reg.plotModelPred(date_list, pred, target)
Reg.plotModelPred(date_list, pred, target, startIndex=200, endIndex=300)
```

```python
# Some statistics of this rather decent prediction!
import matplotlib.mlab as mlab

lag = 1

# error of regression model
err = np.array(pred[lag:]) - np.array(target[lag:])
# error of "baseline model" (predicted temp same as temp on prediction day)
err_base = np.array(target[:-lag]) - np.array(target[lag:])

# histogram of day-over-day changes in daily max temperature
n, bins_base, patches = plt.hist(err_base, 30, normed=1, facecolor='blue', alpha=0.8)
# histogram of errors by prediction model
n, bins, patches = plt.hist(err, 30, normed=1, facecolor='green', alpha=0.5)

# add a 'best fit' normal distribution curves
mu = np.mean(err)
mu_base = np.mean(err_base)
sigma = np.std(err)
sigma_base = np.std(err_base)
y = mlab.normpdf(bins, mu, sigma)
l = plt.plot(bins, y, 'r--', linewidth=2)
y_base = mlab.normpdf(bins_base, mu_base, sigma_base)
l_base = plt.plot(bins_base, y_base, 'r', linewidth=2)

plt.xlabel('error in predicted maximum Temperature')
plt.title('Baseline (blue): mean = %.3f, sigma = %.2f; Reg (green): mean = %.3f, sigma= %.2f' \
          % (mu_base, sigma_base, mu, sigma))
plt.xlim([-18,18])
plt.grid(True)
plt.show()

# Extreme events:
print("Performance when change from prediction day is larger:\n")
for ii in range(10):
     extreme = [(err[i],err_base[i]) for i in range(len(err)) \
                if (abs(err_base[i]) > ii*sigma_base) & (abs(err_base[i]) < (ii+1)*sigma_base)]
     better = [int(e[0] < e[1]) for e in extreme]
     print('There were %d %d-sigma events!' % (len(extreme),ii))
     if (len(extreme) > 0):
          print('Regression beats baseline %.1f%% of the time\n' \
                % (float(100*sum(better)) / float(len(better))))
     else:
          break
```
