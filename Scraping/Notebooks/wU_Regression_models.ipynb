```python
# regression model using data from one city on a single day
# to predict maximum temperature the next day
featureData, target, regr, model_params = Reg.oneCityModel('CYYZ','2005-01-01','2010-12-31',\
                                                           ['TempMax','dailyTempRange',\
                                                            'PressMax','dailyPressRange',\
                                                            'HumidityMean',\
                                                            'isWesterly','isNortherly',\
                                                            'isFoggy'],targetVar='TempMax',lag=1)
```

```python
# out-of-sample prediction
date_list, pred, target = Reg.oneCityPredict(regr, model_params, \
                                             '2011-01-01', '2012-12-31', \
                                             actual=True)
```

```python
# plot comparison
Reg.plotModelPred(date_list, pred, target)
Reg.plotModelPred(date_list, pred, target, startIndex=24, endIndex=85)
```

```python
# using data from one city on multiple days ("Taylor model")
pred0 = pred[2:]  # save simple model prediction (for comparison plot)
featureData, target, regr, model_params = Reg.oneCityTaylorModel('CYYZ','2005-01-01','2010-12-31',\
                                                                 ['TempMax','dailyTempRange',\
                                                                  'PressMax','dailyPressRange',\
                                                                  'HumidityMean',\
                                                                  'isWesterly','isNortherly',\
                                                                  'isFoggy'],targetVar='TempMax',lag=1,order=2)
```

```python
# out-of-sample prediction
date_list, pred, target = Reg.oneCityTaylorPredict(regr, model_params, \
                                                   '2011-01-01', '2012-12-31', \
                                                   actual=True)
```

```python
# plot comparison
Reg.plotModelPred(date_list, pred, target)
Reg.plotModelPred(date_list, pred, target, startIndex=24, endIndex=85)
```
