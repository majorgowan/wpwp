```python
# import Regression module
import pandas as pd

# load Toronto data from CSV file
toronto = pd.read_csv('CSV_DATA/CYYZ.csv', delimiter=', ')
chicago = pd.read_csv('CSV_DATA/KORD.csv', delimiter=', ')
```

```python
toronto.columns
```

```python
# convert date strings to datetime objects (for plotting)
toronto['dateTime'] = [pd.to_datetime(dstr) for dstr in toronto.date]
chicago['dateTime'] = [pd.to_datetime(dstr) for dstr in chicago.date]
```

```python
# take a subset: only 2008 values
tor2008 = toronto[:][(toronto['date'] >= '2008') & (toronto['date'] < '2009')]
chi2008 = chicago[:][(chicago['date'] >= '2008') & (chicago['date'] < '2009')]
```

```python
# plots
import matplotlib.pyplot as plt

fig, ax1 = plt.subplots()
p1tor = ax1.plot(tor2008.dateTime,tor2008.TempMax, color='blue')
p1chi = ax1.plot(chi2008.dateTime,chi2008.TempMax, color='red')
ax1.set_ylabel('Tmax (C)')
ax1.set_ylim([-40,40])
ax1.legend(('Tor','Chi'))
ax1.set_title('Tmax - solid; Pmax - dash')

ax2 = ax1.twinx()
p2tor = ax2.plot(tor2008.dateTime,tor2008.PressMax, color='blue', linestyle='--')
p2chi = ax2.plot(chi2008.dateTime,chi2008.PressMax, color='red', linestyle='--')
ax2.set_ylabel('Pmax (hPa)')
ax2.set_ylim([995,1080])

fig.autofmt_xdate()

plt.show()
```

```python
# scatter plot of next-day Tmax in Toronto vs other variables
fig, ((ax11, ax12, ax13, ax14, ax15), \
      (ax21, ax22, ax23, ax24, ax25)) = \
                  plt.subplots(2, 5, sharex='col', sharey='row')

ax11.scatter(toronto['TempMax'][:-1], toronto['TempMax'][1:])
ax11.set_ylabel('next-day Tor Tmax')
ax11.set_xlabel('Tor Tmax')
ax11.set_xticklabels([])

ax12.scatter(toronto['PressMax'][:-1], toronto['TempMax'][1:])
ax12.set_xlabel('Tor Pmax')
ax12.set_xticklabels([])

ax13.scatter(toronto['HumidityMean'][:-1], toronto['TempMax'][1:])
ax13.set_xlabel('Tor Humid')
ax13.set_xticklabels([])

ax14.scatter(toronto['WindMeanX'][:-1], toronto['TempMax'][1:])
ax14.set_xlabel('Tor WX')
ax14.set_xticklabels([])

ax15.scatter(toronto['WindMeanY'][:-1], toronto['TempMax'][1:])
ax15.set_xlabel('Tor WY')
ax15.set_xticklabels([])

ax21.scatter(chicago['TempMax'][:-1], toronto['TempMax'][1:])
ax21.set_ylabel('next-day Tor Tmax')
ax21.set_xlabel('Chi Tmax')
ax21.set_xticklabels([])

ax22.scatter(chicago['PressMax'][:-1], toronto['TempMax'][1:])
ax22.set_xlabel('Chi Pmax')
ax22.set_xticklabels([])

ax23.scatter(chicago['HumidityMean'][:-1], toronto['TempMax'][1:])
ax23.set_xlabel('Chi Humid')
ax23.set_xticklabels([])

ax24.scatter(chicago['WindMeanX'][:-1], toronto['TempMax'][1:])
ax24.set_xlabel('Chi WX')
ax24.set_xticklabels([])

ax25.scatter(chicago['WindMeanY'][:-1], toronto['TempMax'][1:])
ax25.set_xlabel('Chi WY')
ax25.set_xticklabels([])

plt.show()
```
