```python
# read a month of hourly data from JSON file
import wUnderground as wU
dates, torNov2007 = wU.getJSON('CYYZ',2007,11)
# show variables present (dictionary keys)
torNov2007[0][0].keys()
```

```python
# convert DateUTC to python datetime objects
import datetime
times = [hour['DateUTC'] for day in torNov2007 for hour in day]
times[:5]
```

```python
dtimes = [datetime.datetime.strptime(x,'%Y-%m-%d %H:%M:%S') for x in times]
dtimes[:5]
```

```python
# extract temperature observations as floats
tempC = [float(hour['TemperatureC']) for day in torNov2007 for hour in day]
tempC[:5]
```

```python
# plot temperature versus date
import matplotlib.pyplot as plt
fig = plt.figure()
ax = plt.subplot(111)
plt.plot(dtimes, tempC, color='blue')
plt.ylabel('Temperature (C)')
plt.title('Hourly Temperature for Nov2007 at station CYYZ')
plt.show()
```

```python
# repeat for sea-level pressure
ptimes = [datetime.datetime.strptime(x,'%Y-%m-%d %H:%M:%S') for x in times]
slp = [hour['Sea Level PressurehPa'] for day in torNov2007 for hour in day]
zip(ptimes[:5],slp[:5])
```

```python
# remove missing values (from back to front)
for i in range(len(slp)-1,-1,-1):
    if slp[i] in ['-9999', '-9999.0', 'N/A', ' ', '']:
        del ptimes[i]
        del slp[i]
zip(ptimes[:5],slp[:5])
```

```python
# convert slp values to float and plot
slp = [float(s) for s in slp]
plt.plot(ptimes, slp, color='red')
plt.title('Hourly SLP for Nov2007 at station CYYZ')
plt.ylabel('Sea level pressure (hPa)')
plt.show()
```

```python
# wind speed and direction
wtimes = [datetime.datetime.strptime(x,'%Y-%m-%d %H:%M:%S') for x in times]
wdir = [hour['WindDirDegrees'] for day in torNov2007 for hour in day]
wspd = [hour['Wind SpeedKm/h'] for day in torNov2007 for hour in day]

# remove missing values (from back to front)
for i in range(len(wdir)-1,-1,-1):
    if wdir[i] in ['-9999', 'N/A', ' ', ''] or wspd[i] == 'Calm':
        del wtimes[i]
        del wspd[i]
        del wdir[i]

# convert wind values to radians and plot
wdir = [float(w) for w in wdir]
wspd = [float(wspd) for wspd in wspd]
wdirRad = [float(w)*3.14/180.0 for w in wdir]
wtimes = [(w-wtimes[0]).total_seconds()/24.0/3600.0 for w in wtimes]

fig = plt.figure(figsize=(18.0,6.0))

ax1 = plt.subplot(121, projection='polar')
ax1.set_theta_zero_location("N")
ax1.set_theta_direction(-1)
ax1.set_ylim([0,45.0])
ax1.set_yticks([10.0, 20.0, 30.0, 40.0])
ax1.set_yticklabels(ax1.get_yticks(), fontweight='bold')
ax1.set_title('Observed wind speed and direction at CYYZ in Nov2007')
plt.scatter(wdirRad, wspd, color='steelblue', alpha=0.4, s=30)

ax2 = plt.subplot(122)
n, bins, patches = plt.hist(wdir, 18, normed=1, facecolor='plum', alpha=0.8)
ax2.set_title('Distribution of wind directions observed at CYYZ in Nov2007')
ax2.set_xlim([0,360])
ax2.set_xlabel('Wind Direction (degrees east of north)')
plt.show()
```

```python
# compute daily average (R = rectangle method; T = trapezoid method)
import wUStats as Stats
TmeanR, TmeanT = Stats.dailyMean(torNov2007, 'TemperatureC', 'rectangle'), \
                 Stats.dailyMean(torNov2007, 'TemperatureC', 'trapezoid')
PmeanR, PmeanT = Stats.dailyMean(torNov2007, 'Sea Level PressurehPa', 'rectangle'), \
                 Stats.dailyMean(torNov2007, 'Sea Level PressurehPa', 'trapezoid')
```

```python
# plot daily means
fig, (axT, axP) = plt.subplots(2, sharex=True)

axT.plot(dates,TmeanR,color='red')
axT.plot(dates,TmeanT,color='blue')
axT.set_ylabel('Temp (C)')

axP.plot(dates,PmeanR,color='red')
axP.plot(dates,PmeanT,color='blue')
axP.set_ylabel('SLP (hPa)')

fig.autofmt_xdate()

plt.show()
```

```python
# scatter plots of next-day daily max temperature with
# daily max temperature, daily mean pressure, daily max pressure, daily max wind speed
Tmax, Tmaxpos = Stats.dailyMax(torNov2007, 'TemperatureC', minmax=1)
Tnext = Tmax[1:]
Ttoday = Tmax[:-1]

Pmean = Stats.dailyMean(torNov2007, 'Sea Level PressurehPa', 'rectangle')
Pmean = Pmean[:-1]

Pmax, Pmaxpos = Stats.dailyMax(torNov2007, 'Sea Level PressurehPa')
Pmax = Pmax[:-1]

Wmax, Wmaxdir, Wmaxpos = Stats.dailyMaxWind(torNov2007)
Wmax = Wmax[:-1]

# plots
fig, (axTmax, axPmean, axPmax, axWmax) = plt.subplots(1, 4, sharey=True)

axTmax.scatter(Ttoday, Tnext)
axTmax.set_ylabel('Next-day Tmax (C)')
axTmax.set_xlabel('Tmax (C)')
axTmax.set_xticklabels([])

axPmean.scatter(Pmean, Tnext)
axPmean.set_xlabel('Pmean (hPa)')
axPmean.set_xticklabels([])

axPmax.scatter(Pmax, Tnext)
axPmax.set_xlabel('Pmax (hPa)')
axPmax.set_xticklabels([])

axWmax.scatter(Wmax, Tnext)
axWmax.set_xlabel('Wmax (kph)')
axWmax.set_xticklabels([])

plt.show()
```
