### A Regression model

```python
# build a regression model using only Toronto data
import statsmodels.formula.api as sm

tor_lag = toronto[:][:-1]
tor_lag['Tmax_next'] = [toronto['TempMax'][i] for i in range(1,len(toronto))]

# compute some derived variables
tor_lag['Trange'] = [tor_lag['TempMax'][i]-tor_lag['TempMin'][i] for i in range(len(tor_lag))]
tor_lag['Prange'] = [tor_lag['PressMax'][i]-tor_lag['PressMin'][i] for i in range(len(tor_lag))]

# split into training and testing
train = tor_lag[:][(tor_lag['date'] < '2011')]
test = tor_lag[:][(tor_lag['date'] >= '2011')]

tor_form = "Tmax_next ~ TempMax + Trange + PressMax + " \
           + "Prange + HumidityMean + VisibilityMean + " \
           + "WindMaxSpd + WindMeanX + WindMeanY"

tor_model = sm.ols(formula = tor_form, data=train).fit()

print(tor_model.summary())
```

```python
# correlations between features in dataframe
tor_lag.corr()
```

```python
# make predictions on testing set
import numpy as np

pred = tor_model.predict(test)
sse = sum((np.array(pred) - np.array(test['Tmax_next']))**2)
sse_base = sum((np.array(test['TempMax']) - np.array(test['Tmax_next']))**2)
Rsq = 1 - sse/sse_base
rmse = np.sqrt(sse/len(test))
print('rmse:\t' + str(rmse))
print('Rsq:\t' + str(Rsq))
```

```python
# build a regression model using Toronto AND Chicago data
import statsmodels.formula.api as sm

tor_lag = toronto[:][:-1]
chi_lag = chicago[:][:-1]

# compute some derived variables
tor_lag['Trange'] = [tor_lag['TempMax'][i]-tor_lag['TempMin'][i] for i in range(len(tor_lag))]
tor_lag['Prange'] = [tor_lag['PressMax'][i]-tor_lag['PressMin'][i] for i in range(len(tor_lag))]
chi_lag['Trange'] = [chi_lag['TempMax'][i]-chi_lag['TempMin'][i] for i in range(len(chi_lag))]
chi_lag['Prange'] = [chi_lag['PressMax'][i]-chi_lag['PressMin'][i] for i in range(len(chi_lag))]

# rename columns in both dataframes
names = tor_lag.columns.values
names = [x + '_tor' for x in names]
tor_lag.columns = names
names = chi_lag.columns.values
names = [x + '_chi' for x in names]
chi_lag.columns = names

# add next-day temperature in Toronto
tor_lag['Tmax_next'] = [toronto['TempMax'][i] for i in range(1,len(toronto))]

# combine two dataframes
tor_lag = pd.concat([tor_lag, chi_lag], axis=1)

# split into training and testing
train = tor_lag[:][(tor_lag['date_tor'] < '2011')]
test = tor_lag[:][(tor_lag['date_tor'] >= '2011')]

tor_form = "Tmax_next ~ TempMax_tor + Trange_tor + PressMax_tor + " \
           + "Prange_tor + HumidityMean_tor + VisibilityMean_tor + " \
           + "WindMaxSpd_tor + WindMeanX_tor + WindMeanY_tor + " \
           + "TempMax_chi + Trange_chi + PressMax_chi + " \
           + "Prange_chi + HumidityMean_chi + VisibilityMean_chi + " \
           + "WindMaxSpd_chi + WindMeanX_chi + WindMeanY_chi"

tor_model = sm.ols(formula = tor_form, data=train).fit()

print(tor_model.summary())
```

```python
# predictions
pred = tor_model.predict(test)
sse = sum((np.array(pred) - np.array(test['Tmax_next']))**2)
sse_base = sum((np.array(test['TempMax_tor']) - np.array(test['Tmax_next']))**2)
Rsq = 1 - sse/sse_base
rmse = np.sqrt(sse/len(test))
print('rmse:\t' + str(rmse))
print('Rsq:\t' + str(Rsq))
```

```python
# using sklearn instead

import numpy as np
from sklearn import linear_model
tor_lag[['Tmax_next','TempMax_tor']][:3]

features = train[['TempMax_tor', 'Trange_tor', 'PressMax_tor', \
                  'Prange_tor', 'HumidityMean_tor', 'VisibilityMean_tor', \
                  'WindMaxSpd_tor', 'WindMeanX_tor', 'WindMeanY_tor', \
                  'TempMax_chi', 'Trange_chi', 'PressMax_chi', \
                  'Prange_chi', 'HumidityMean_chi', 'VisibilityMean_chi', \
                  'WindMaxSpd_chi', 'WindMeanX_chi', 'WindMeanY_chi']]
target = train[['Tmax_next']]

regr = linear_model.LinearRegression().fit(features, target)
print("R^2: " + str(regr.score(features,target)))
print("Regression coefficients:")
print("  intercept" + ":\t" + str(regr.intercept_[0]))
for ii, feature in enumerate(features.columns):
    print("  " + feature + ":\t" + str(regr.coef_[0][ii]))
```

```python
# predictions on the test set
features_test = test[['TempMax_tor', 'Trange_tor', 'PressMax_tor', \
                      'Prange_tor', 'HumidityMean_tor', 'VisibilityMean_tor', \
                      'WindMaxSpd_tor', 'WindMeanX_tor', 'WindMeanY_tor', \
                      'TempMax_chi', 'Trange_chi', 'PressMax_chi', \
                      'Prange_chi', 'HumidityMean_chi', 'VisibilityMean_chi', \
                      'WindMaxSpd_chi', 'WindMeanX_chi', 'WindMeanY_chi']]
target_test = test[['Tmax_next']]

pred = regr.predict(features_test)
sse = sum((np.array(pred) - np.array(target_test))**2)
sse_base = sum((np.array(test['TempMax_tor']) - np.array(test['Tmax_next']))**2)
Rsq = 1 - sse/sse_base
rmse = np.sqrt(sse/len(test))
print('rmse:\t' + str(rmse))
print('Rsq:\t' + str(Rsq))
```

```python
# stepwise feature selection
from sklearn.feature_selection import RFE

regr = linear_model.LinearRegression()

rfe = RFE(regr)
rfe.fit(features, train['Tmax_next'])

best_features = features.columns.values[rfe.support_]
print('Best features: ' + str(best_features))

features_new = rfe.transform(features)
regr.fit(features_new, target)
print("R^2: " + str(regr.score(features_new,target)))
print("Regression coefficients:")
print("  intercept" + ":\t" + str(regr.intercept_[0]))
for ii, feature in enumerate(best_features):
    print("  " + feature + ":\t" + str(regr.coef_[0][ii]))
```

```python
# . . . predictions
features_test_new = rfe.transform(features_test)
target_test = test[['Tmax_next']]

pred = regr.predict(features_test_new)
sse = sum((np.array(pred) - np.array(target_test))**2)
sse_base = sum((np.array(test['TempMax_tor']) - np.array(test['Tmax_next']))**2)
Rsq = 1 - sse/sse_base
rmse = np.sqrt(sse/len(test))
rmse_base = np.sqrt(sse_base/len(test))
print('rmse:\t' + str(rmse) + '  ( base:\t' + str(rmse_base) + ' )')
print('Rsq:\t' + str(Rsq))
```

```python
# some more statistics views
import matplotlib.mlab as mlab

err = np.array(pred) - np.array(target_test)
err_base = np.array(test['TempMax_tor']) - np.array(test['Tmax_next'])

figure(figsize=(7.0, 4.0))
# histogram of errors by prediction model
n, bins, patches = plt.hist(err, 30, normed=1, facecolor='green', alpha=0.75)

# add a 'best fit' line
mu = np.mean(err)
sigma = np.std(err)
y = mlab.normpdf(bins, mu, sigma)
l = plt.plot(bins, y, 'r', linewidth=2)

plt.xlabel('prediction error')
plt.title('Histogram, mean = %f, sigma = %f' % (mu, sigma))
plt.xlim([-18,18])
plt.grid(True)
plt.show()

figure(figsize=(7.0, 4.0))
# histogram of day-over-day changes in daily max temperature
n, bins_base, patches = plt.hist(err_base, 30, normed=1, facecolor='blue', alpha=0.75)

# add a 'best fit' line
mu_base = np.mean(err_base)
sigma_base = np.std(err_base)
y_base = mlab.normpdf(bins_base, mu_base, sigma_base)
l_base = plt.plot(bins_base, y_base, 'r', linewidth=2)
l = plt.plot(bins, y, 'r--', linewidth=2)

plt.xlabel('day-over-day change in daily Tmax')
plt.title('Histogram, mean = %f, sigma = %f' % (mu_base, sigma_base))
plt.xlim([-18,18])
plt.grid(True)
plt.show()
```

```python
# how often does regression beat baseline?
better = [int(abs(err[i]) < abs(err_base[i])) for i in range(len(err))]
print('Regression beats baseline %.1f%% of the time' % (float(100*sum(better)) / float(len(better))))
```

```python
# extreme events (day-over-day change more than 1-sigma)
extreme = [(err[i],err_base[i]) for i in range(len(err)) if abs(err_base[i]) > sigma_base]
extreme2 = [(err[i],err_base[i]) for i in range(len(err)) if abs(err_base[i]) > 2*sigma_base]
extreme3 = [(err[i],err_base[i]) for i in range(len(err)) if abs(err_base[i]) > 3*sigma_base]

print('There were %d total events, of which . . .\n' % len(err))

better_extreme = [int(abs(e[0]) < abs(e[1])) for e in extreme]
print('There were %d 1-sigma events!' % len(extreme))
print('In 1-sigma events, regression beats baseline %.1f%% of the time\n' \
      % (float(100*sum(better_extreme)) / float(len(better_extreme))))

better_extreme = [int(abs(e[0]) < abs(e[1])) for e in extreme2]
print('There were %d 2-sigma events!' % len(extreme2))
print('In 2-sigma events, regression beats baseline %.1f%% of the time\n' \
      % (float(100*sum(better_extreme)) / float(len(better_extreme))))

better_extreme = [int(abs(e[0]) < abs(e[1])) for e in extreme3]
print('There were %d 3-sigma events!' % len(extreme3))
print('In 3-sigma events, regression beats baseline %.1f%% of the time\n' \
      % (float(100*sum(better_extreme)) / float(len(better_extreme))))
```
