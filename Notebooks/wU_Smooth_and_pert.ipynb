```python
import wUUtils as Util
toronto = Util.loadDailyVariableRange('CYYZ','2005-01-01','2010-12-31','TempMax',castFloat=True)
newyork = Util.loadDailyVariableRange('KJFK','2005-01-01','2010-12-31','TempMax',castFloat=True)
desmoines = Util.loadDailyVariableRange('KDSM','2005-01-01','2010-12-31','TempMax',castFloat=True)
```

```python
import numpy as np

toronto_smooth = Util.smooth(toronto,90)
newyork_smooth = Util.smooth(newyork,90)
desmoines_smooth = Util.smooth(desmoines,90)

toronto_pert = np.array(toronto)-np.array(toronto_smooth)
newyork_pert = np.array(newyork)-np.array(newyork_smooth)
desmoines_pert = np.array(desmoines)-np.array(desmoines_smooth)

date_list = Util.dateList('2005-01-01','2010-12-31')
```

```python
def sigmas(station,data,data_smooth,data_pert):
    print('%s: \t sigma: %.2f \t sigma_smooth: %.2f \t sigma_pert: %.2f' \
                % (station,np.std(data),np.std(data_smooth),np.std(data_pert)))

sigmas('toronto',toronto,toronto_smooth,toronto_pert)
sigmas('new york',newyork,newyork_smooth,newyork_pert)
sigmas('des moines',desmoines,desmoines_smooth,desmoines_pert)
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(14,5))
plt.plot(date_list,toronto, color='blue', alpha=0.6)
plt.plot(date_list,toronto_smooth, color='red', alpha=0.3, linewidth=8)
plt.title('Toronto')
plt.show()

fig = plt.figure(figsize=(14,5))
plt.plot(date_list,toronto_smooth, color='blue')
plt.plot(date_list,newyork_smooth, color='red')
plt.plot(date_list,desmoines_smooth, color='green')
plt.legend(['Toronto','New York','Des Moines'], fontsize=10, ncol=3)
plt.title('90-day running mean MaxTemp')
plt.ylim([-5,35])
plt.show()

fig = plt.figure(figsize=(14,5))
plt.plot(date_list,toronto_pert, color='blue', alpha=0.6)
plt.plot(date_list,newyork_pert, color='red', alpha=0.6)
plt.plot(date_list,desmoines_pert, color='green', alpha=0.6)
plt.legend(['Toronto','New York','Des Moines'], fontsize=10, ncol=3)
plt.title('Departure from 90-day running mean MaxTemp')
plt.ylim([-20,20])
plt.show()
```

```python
# climatology
startDate, endDate = '2005-01-01', '2010-12-31'

one_year = Util.dateList('2005-01-01','2005-12-31')
date_list = Util.dateList(startDate,endDate)

variable = 'TempMax'

climatology=[]
stations = Util.getStationList()
data = Util.loadDailyVariableSetRange(stations,startDate,endDate,['TempMax'])

for day_of_month in one_year:
    this_day = mean([stationData[i] for i in range(len(date_list)) \
                     for stationData in data \
                     if date_list[i].month==day_of_month.month \
                     and date_list[i].day==day_of_month.day])
    climatology.append(this_day)
climatology = Util.smooth(climatology,5)
```

```python
fig = plt.figure(figsize=(10.0,10.0))
plt.plot(one_year,climatology,alpha=0.75,linewidth=3)
plt.plot(one_year,toronto[:365],color='red',alpha=0.5,linewidth=3)
plt.plot(one_year,desmoines[-365:],color='green',alpha=0.5,linewidth=3)
plt.title('All-Station 2005-2010 Climatology TempMax',size=14)
plt.legend(['Climatology','Toronto 2005','Des Moines 2010'],fontsize=14)
fig.autofmt_xdate()
plt.grid()
plt.show()
```
