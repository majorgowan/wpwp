```python
import wUUtils as Util
toronto = Util.loadDailyVariableRange('CYYZ','2005-01-01','2010-12-31','TempMax',castFloat=True)
newyork = Util.loadDailyVariableRange('KJFK','2005-01-01','2010-12-31','TempMax',castFloat=True)
desmoines = Util.loadDailyVariableRange('KDSM','2005-01-01','2010-12-31','TempMax',castFloat=True)
```

```python
import numpy as np

toronto_smooth = Util.smooth(toronto,90)
newyork_smooth = Util.smooth(newyork,90)
desmoines_smooth = Util.smooth(desmoines,90)

toronto_pert = np.array(toronto)-np.array(toronto_smooth)
newyork_pert = np.array(newyork)-np.array(newyork_smooth)
desmoines_pert = np.array(desmoines)-np.array(desmoines_smooth)

date_list = Util.dateList('2005-01-01','2010-12-31')
```

```python
def sigmas(station,data,data_smooth,data_pert):
    print('%s: \t sigma: %.2f \t sigma_smooth: %.2f \t sigma_pert: %.2f' \
                % (station,np.std(data),np.std(data_smooth),np.std(data_pert)))

sigmas('toronto',toronto,toronto_smooth,toronto_pert)
sigmas('new york',newyork,newyork_smooth,newyork_pert)
sigmas('des moines',desmoines,desmoines_smooth,desmoines_pert)
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(14,5))
plt.plot(date_list,toronto, color='blue', alpha=0.6)
plt.plot(date_list,toronto_smooth, color='red', alpha=0.3, linewidth=8)
plt.title('Toronto')
plt.show()

fig = plt.figure(figsize=(14,5))
plt.plot(date_list,toronto_smooth, color='blue')
plt.plot(date_list,newyork_smooth, color='red')
plt.plot(date_list,desmoines_smooth, color='green')
plt.legend(['Toronto','New York','Des Moines'], fontsize=10, ncol=3)
plt.title('90-day running mean MaxTemp')
plt.ylim([-5,35])
plt.show()

fig = plt.figure(figsize=(14,5))
plt.plot(date_list,toronto_pert, color='blue', alpha=0.6)
plt.plot(date_list,newyork_pert, color='red', alpha=0.6)
plt.plot(date_list,desmoines_pert, color='green', alpha=0.6)
plt.legend(['Toronto','New York','Des Moines'], fontsize=10, ncol=3)
plt.title('Departure from 90-day running mean MaxTemp')
plt.ylim([-20,20])
plt.show()
```

```python
fig = plt.figure(figsize=(20.0,5.0))

# add a 'best fit' line
def plotHisto(data):
    mu = np.mean(data)
    sigma = np.std(data)
    n, bins, patches = plt.hist(data, 30, normed=1, facecolor='blue', alpha=0.75)
    y = mlab.normpdf(bins, mu, sigma)
    fitLine = plt.plot(bins, y, 'r', linewidth=2)
    l = plt.plot(bins, y, 'r--', linewidth=2)
    plt.grid()
    return mu, sigma

ax1 = plt.subplot(1,3,1)
mu_tor, sigma_tor = plotHisto(toronto_pert)
ax1.set_title('Toronto; mu=%.2f; sigma=%.2f' % (mu_tor, sigma_tor))

ax2 = plt.subplot(1,3,2)
mu_ny, sigma_ny = plotHisto(newyork_pert)
ax2.set_title('New York; mu=%.2f; sigma=%.2f' % (mu_ny, sigma_ny))
    
ax3 = plt.subplot(1,3,3)
mu_dm, sigma_dm = plotHisto(desmoines_pert)
ax3.set_title('Des Moines; mu=%.2f; sigma=%.2f' % (mu_dm, sigma_dm))

plt.show()
```

```python
# climatology
startDate, endDate = '2005-01-01', '2010-12-31'

one_year = Util.dateList('2005-01-01','2005-12-31')
date_list = Util.dateList(startDate,endDate)

variable = 'TempMax'

climatology=[]
stations = Util.getStationList()
data = Util.loadDailyVariableSetRange(stations,startDate,endDate,['TempMax'])

for day_of_month in one_year:
    this_day = mean([stationData[i] for i in range(len(date_list)) \
                     for stationData in data \
                     if date_list[i].month==day_of_month.month \
                     and date_list[i].day==day_of_month.day])
    climatology.append(this_day)
climatology = Util.smooth(climatology,5)
```

```python
# Statistics of climatology and no-climatology models
import matplotlib.mlab as mlab

lag=1

# error of regression model
err = np.array(pred[lag:]) - np.array(target[lag:])
# error of no-climatology
err_noc = np.array(pred_old[lag:]) - np.array(target_old[lag:])

# error of "baseline model" (predicted temp same as temp on prediction day)
err_base = np.array(target[:-lag]) - np.array(target[lag:])
sigma_base = np.std(err_base)

fig = plt.figure(figsize=(10.0,6.0))

# histogram of day-over-day changes in daily max temperature
n, bins_noc, patches = plt.hist(err_noc, 30, normed=1, facecolor='blue', alpha=0.8)
# histogram of errors by prediction model
n, bins, patches = plt.hist(err, 30, normed=1, facecolor='green', alpha=0.5)

# add a 'best fit' normal distribution curves
mu = np.mean(err)
mu_noc = np.mean(err_noc)
sigma = np.std(err)
sigma_noc = np.std(err_noc)
y = mlab.normpdf(bins, mu, sigma)
l = plt.plot(bins, y, 'r--', linewidth=2)
y_noc = mlab.normpdf(bins_noc, mu_noc, sigma_noc)
l_noc = plt.plot(bins_noc, y_noc, 'r', linewidth=2)

plt.xlabel('error in predicted maximum Temperature')
plt.title('No-climatology (blue): mean = %.3f, sigma = %.2f; With-climatology (green): mean = %.3f, sigma= %.2f' \
          % (mu_noc, sigma_noc, mu, sigma))
plt.xlim([-18,18])
plt.grid(True)
plt.show()

# Extreme events:
print("Performance when change from prediction day is larger:\n")
for ii in range(10):
     extreme = [(err[i],err_noc[i],err_base[i]) for i in range(len(err)) \
                if (abs(err_base[i]) > ii*sigma_base) & (abs(err_base[i]) < (ii+1)*sigma_base)]
     better = [int(abs(e[0]) < abs(e[1])) for e in extreme]
     print('There were %d %d-sigma events!' % (len(extreme),ii))
     if (len(extreme) > 0):
          print('With-climatology beats no-climatology %.1f%% of the time' \
                % (float(100*sum(better)) / float(len(better))))
          rmse = np.sqrt(np.mean([e[1]*e[1] for e in extreme]))
          print('No-climatology RMSE : \t%.2f' % rmse)
          rmse = np.sqrt(np.mean([e[0]*e[0] for e in extreme]))
          print('With-climatology RMSE : \t%.2f\n' % rmse)
     else:
          break
```

```python
# combination model: use climatology for large predicted changes (actually bad idea: no-climatology model always better)
pred_hybrid = [pred_old[0]]
for i in range(1,len(pred_abso)):
    if abs(pred_old[i]-target_old[i-1]) < 7:
        pred_hybrid.append(pred_old[i])
    else:
        pred_hybrid.append(pred_abso[i])

# error of hybrid
err_hybrid = np.array(pred_hybrid[lag:]) - np.array(target_old[lag:])
rmse_hybrid = np.sqrt(np.mean([e**2 for e in err_hybrid]))
print('Hybrid model: rmse %.2f' %rmse_hybrid)
```

### Up-or-down accuracy

```python
target_updown = [int(target_old[i]-target_old[i-1] > 0) for i in range(1,len(target_old))]
pred_updown = [int(pred_abso[i]-target_old[i-1] > 0) for i in range(1,len(target_old))]
pred_old_updown = [int(pred_old[i]-target_old[i-1] > 0) for i in range(1,len(target_old))]

accuracy = sum(int(pred_updown[i]==target_updown[i]) for i in range(len(target_updown)))
accuracy_old = sum(int(pred_old_updown[i]==target_updown[i]) for i in range(len(target_updown)))
```

```python
print('Up-down accuracy with climatology: %.2f' % (float(accuracy)/len(target_updown)))
print('Up-down accuracy without climatology: %.2f' % (float(accuracy_old)/len(target_updown)))
```

```python
def contingency(A,B):
    import numpy as np
    zipper = zip(A,B)
    FF = sum(int(not z[0] and not z[1]) for z in zipper)
    FT = sum(int(not z[0] and z[1]) for z in zipper)
    TF = sum(int(z[0] and not z[1]) for z in zipper)
    TT = sum(int(z[0] and z[1]) for z in zipper)
    return np.array([[FF, FT],[TF, TT]])

confusion = contingency(target_updown,pred_old_updown)
print('Confusion matrix')
print(confusion)

print('\n accuracy: %.1f%%' % (float(100.0*np.trace(confusion))/sum(confusion)))
```
