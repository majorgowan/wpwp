### Slide 4 - DataSet

```python
# read a month of hourly data from JSON file
import wUnderground as wU
dates, torNov2007 = wU.getJSON('CYYZ',2007,11)

# convert DateUTC to python datetime objects
import datetime
times = [hour['DateUTC'] for day in torNov2007 for hour in day]
dtimes = [datetime.datetime.strptime(x,'%Y-%m-%d %H:%M:%S') for x in times]

# extract temperature observations as floats
tempC = [float(hour['TemperatureC']) for day in torNov2007 for hour in day]

# plot temperature versus date
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(4.5,3.5))
ax = plt.subplot(111)
plt.plot(dtimes, tempC, color='blue')
plt.ylabel('Temperature (C)',fontsize=14,color='darkred')
plt.title('Hourly Temp at CYYZ',fontsize=16,color='darkred')
fig.autofmt_xdate()

ax.set_yticklabels(ax.get_yticks(), fontsize=12,color='darkred')
for t in ax.get_xticklabels():
    t.set_fontsize(10)
    t.set_color('darkred')
ax.grid()
plt.show()
```

```python
# PCA models
# using data from multiple cities on multiple days
import wUClusterRegression as ClustReg

# get list of all available stations
import wUUtils as Util
stations = Util.getStationList()
# move target station to front
targStation = 'CYYZ'
stations = [targStation] + [s for s in stations if s != targStation]
# build model
nclust = 4
featureData, target, modelParams = ClustReg.pcaClusterModel(stations,'2005-01-01','2010-12-31',\
                                                            ['TempMax','dailyTempRange','PressMax','dailyPressRange',\
                                                            'HumidityMean', \
                                                            'isSoutherly','isWesterly','isFoggy'], \
                                                            clusterVars=[('TempMax',2),('PressMax',2)], nclusters=nclust, \
                                                            ncomp=[3, 3, 3, 3, 3, 3, 3, 3], \
                                                            targetVar='TempMax',lag=1,order=0,verbose=False)

# make predictions
date_list, pred, target, \
     pred_featureData, pred_classes, modelPerf = ClustReg.pcaClusterPredict(modelParams, \
                                                                            '2011-01-01', '2012-12-31', \
                                                                            actual=True)
    
# plot test set values colour-coded by cluster
colors = ['blue','red','green','black','cyan','yellow']
fig = plt.figure(figsize=(8.0,5.0))
ax = plt.subplot(111)
ax.set_axis_bgcolor('lavender')
ax.grid()
ax.set_ylabel('Temperature (C)')
ax.set_title('Daily maximum temperature at CYYZ',size=14)
legs = []
for icl in range(nclust):
     cl0 = [t for i,t in enumerate(target) if pred_classes[i]==icl]
     dt0 = [t for i,t in enumerate(date_list) if pred_classes[i]==icl]
     plt.scatter(dt0,cl0,color=colors[icl])
     legs.append('RMSE=%.2f' % modelPerf['RMSE'][icl])
plt.legend(legs,fontsize=10)
fig.autofmt_xdate()
plt.show()
```

```python
# visualize clusters
clusterColumns = [2,8]
train_data0 = featureData[clusterColumns[0]]
train_data1 = featureData[clusterColumns[1]]
test_data0 = pred_featureData[clusterColumns[0]]
test_data1 = pred_featureData[clusterColumns[1]]

train_classes = modelParams['classes']

fig = plt.figure(figsize=(7.0,6.0))
ax = fig.add_subplot(1,1,1)
plt.plot([0,0],[-6,6],color='black',linestyle='--',linewidth=2)
plt.plot([-3,3],[0,0],color='black',linestyle='--',linewidth=2)
ax.set_xlim([-3,3])
ax.set_ylim([-6,6])
legs=[]
scatPlots = []
for icl in range(nclust):
     cl0 = [t for i,t in enumerate(train_data0) if train_classes[i]==icl]
     cl1 = [t for i,t in enumerate(train_data1) if train_classes[i]==icl]
     ax.scatter(cl0,cl1,color=colors[icl],alpha=0.2,marker='D')

     cl0 = [t for i,t in enumerate(test_data0) if pred_classes[i]==icl]
     cl1 = [t for i,t in enumerate(test_data1) if pred_classes[i]==icl]
     scatPlots.append(ax.scatter(cl0,cl1,color=colors[icl],alpha=0.6))
     
     legs.append('RMSE=%.1f' % (modelPerf['RMSE'][icl]))

plt.legend(scatPlots,legs,fontsize=13)

ax.grid(True)
ax.set_title('k-Means clusters',size=16)
ax.set_xlabel('Temperature PC2',size=12)
ax.set_ylabel('Pressure PC2',size=12)
plt.show()
```
