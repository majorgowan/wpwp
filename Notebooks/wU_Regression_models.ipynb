### Multi-city, multi-day model

```python
# using data from multiple cities on multiple days

# get list of all available stations
import wUUtils as Util
stations = Util.getStationList()
# move target station to front
targStation = 'CYYZ'
stations = [targStation] + [s for s in stations if s != targStation]
# build model
featureData, target, model_params = Reg.multiCityTaylorModel(stations,'2005-01-01','2010-12-31',\
                                                           ['TempMax','dailyTempRange',\
                                                            'PressMax','dailyPressRange',\
                                                            'HumidityMean',\
                                                            'isWesterly','isSoutherly',\
                                                            'isFoggy'], \
                                                            targetVar='TempMax',lag=1,order=1,scale=True,verbose=False)
```

```python
# map of average (sum) regression coefficients by station:
import wUMapPlots as Map

stations = model_params['stations']
features = model_params['features']
regr = model_params['regr']
order = model_params['order']

nfeat = len(features)
nstat = len(stations)
sum_coef = []
for i,station in enumerate(stations):
    vals = []
    for jj in range(order):
        start = jj*nstat*nfeat + i*nfeat
        end = jj*nstat*nfeat + (i+1)*nfeat
        vals += [abs(coco) for coco in regr.coef_[start:end]]
    sum_coef.append(sum(vals))

Map.contourPlotStationsOnMap(stations, sum_coef, title='sum of scaled regression coefficients', width_fac=18, height_fac=15)
```

```python
# predict
date_list, pred, target, model_perf = Reg.multiCityTaylorPredict(model_params, \
                                                                 '2011-01-01', '2012-12-31', \
                                                                 actual=True)
ax = Reg.plotModelPred(date_list, pred, target, startIndex=200, endIndex=300, showIt=False)
ax.set_title('Multi-city model prediction for CYYZ')
ax.set_ylabel('Temperature (C)')
ax.grid(True)
plt.show()
```

```python
# Some statistics of this rather decent prediction!
import matplotlib.mlab as mlab

lag = 1

# error of regression model
err = np.array(pred[lag:]) - np.array(target[lag:])
# error of "baseline model" (predicted temp same as temp on prediction day)
err_base = np.array(target[:-lag]) - np.array(target[lag:])

fig = plt.figure(figsize=(10.0, 6.0))
# histogram of day-over-day changes in daily max temperature
n, bins_base, patches = plt.hist(err_base, 30, normed=1, facecolor='blue', alpha=0.6)
# histogram of errors by prediction model
n, bins, patches = plt.hist(err, 30, normed=1, facecolor='green', alpha=0.5)

# add a 'best fit' normal distribution curves
mu = np.mean(err)
mu_base = np.mean(err_base)
sigma = np.std(err)
sigma_base = np.std(err_base)
y = mlab.normpdf(bins, mu, sigma)
l = plt.plot(bins, y, 'r--', linewidth=2)
y_base = mlab.normpdf(bins_base, mu_base, sigma_base)
l_base = plt.plot(bins_base, y_base, 'r', linewidth=2)

plt.xlabel('error in predicted maximum Temperature')
plt.title('Baseline (blue): mean = %.3f, sigma = %.2f; Reg (green): mean = %.3f, sigma= %.2f' \
          % (mu_base, sigma_base, mu, sigma))
plt.xlim([-18,18])
plt.grid(True)
plt.show()

# Extreme events:
print("Performance when change from prediction day is larger:\n")
for ii in range(10):
     extreme = [(err[i],err_base[i]) for i in range(len(err)) \
                if (abs(err_base[i]) > ii*sigma_base) & (abs(err_base[i]) < (ii+1)*sigma_base)]
     better = [int(abs(e[0]) < abs(e[1])) for e in extreme]
     print('There were %d %d-sigma events!' % (len(extreme),ii))
     if (len(extreme) > 0):
          print('Regression beats baseline %.1f%% of the time' \
                % (float(100*sum(better)) / float(len(better))))
          rmse = np.sqrt(np.mean([e[1]*e[1] for e in extreme]))
          print('Baseline RMSE : \t%.2f' % rmse)
          rmse = np.sqrt(np.mean([e[0]*e[0] for e in extreme]))
          print('Regression RMSE : \t%.2f\n' % rmse)
     else:
          break
```

```python
# map regression coefficients back to station-space
regr = model_params['regr']
ncomp = model_params['transform_params']['ncomp']
order = model_params['order']
features = model_params['features']
pcas = model_params['transform_params']['pcas']

nfeat = len(features)
station_coefs = []
for jj in range(order+1):
    for ifeat, feature in enumerate(features):
        start = jj*sum(ncomp) + sum(ncomp[:ifeat])
        end = start + ncomp[ifeat]
        station_coefs.append(pcas[ifeat].inverse_transform(regr.coef_[start:end]))
        
Map.contourPlotStationsOnMap(stations, station_coefs[0], title='Temperature Coefficients', width_fac=18, height_fac=15)
Map.contourPlotStationsOnMap(stations, station_coefs[1], title='Temperature Range Coefficients', width_fac=18, height_fac=15)
Map.contourPlotStationsOnMap(stations, station_coefs[8], title='Temperature Tendency Coefficients', width_fac=18, height_fac=15)

Map.contourPlotStationsOnMap(stations, station_coefs[2], title='Pressure Coefficients', width_fac=18, height_fac=15)
Map.contourPlotStationsOnMap(stations, station_coefs[7], title='Fogginess Coefficients', width_fac=18, height_fac=15)
```

```python
# Statistics of boffo-looking average prediction!
import matplotlib.mlab as mlab

lag = 1

# error of regression model
err = np.array(pred_mean[lag:]) - np.array(target[lag:])
# error of "baseline model" (predicted temp same as temp on prediction day)
err_base = np.array(target[:-lag]) - np.array(target[lag:])

fig = plt.figure(figsize=(10.0, 6.0))
# histogram of day-over-day changes in daily max temperature
n, bins_base, patches = plt.hist(err_base, 30, normed=1, facecolor='blue', alpha=0.8)
# histogram of errors by prediction model
n, bins, patches = plt.hist(err, 30, normed=1, facecolor='green', alpha=0.5)

# add a 'best fit' normal distribution curves
mu = np.mean(err)
mu_base = np.mean(err_base)
sigma = np.std(err)
sigma_base = np.std(err_base)
y = mlab.normpdf(bins, mu, sigma)
l = plt.plot(bins, y, 'r--', linewidth=2)
y_base = mlab.normpdf(bins_base, mu_base, sigma_base)
l_base = plt.plot(bins_base, y_base, 'r', linewidth=2)

plt.xlabel('error in predicted maximum Temperature')
plt.title('Baseline (blue): mean = %.3f, sigma = %.2f; Reg (green): mean = %.3f, sigma= %.2f' \
          % (mu_base, sigma_base, mu, sigma))
plt.xlim([-18,18])
plt.grid(True)
plt.show()

# Extreme events:
print("Performance when change from prediction day is larger:\n")
for ii in range(10):
     extreme = [(err[i],err_base[i]) for i in range(len(err)) \
                if (abs(err_base[i]) > ii*sigma_base) & (abs(err_base[i]) < (ii+1)*sigma_base)]
     better = [int(abs(e[0]) < abs(e[1])) for e in extreme]
     print('There were %d %d-sigma events!' % (len(extreme),ii))
     if (len(extreme) > 0):
          print('Regression beats baseline %.1f%% of the time' \
                % (float(100*sum(better)) / float(len(better))))
          rmse = np.sqrt(np.mean([e[1]*e[1] for e in extreme]))
          print('Baseline RMSE : \t%.2f' % rmse)
          rmse = np.sqrt(np.mean([e[0]*e[0] for e in extreme]))
          print('Regression RMSE : \t%.2f\n' % rmse)
     else:
          break
```
